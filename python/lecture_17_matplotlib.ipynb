#Matplotlib

##`pylab` and `pyplot`

The `matplotlib.pyplot` module contains the API for functions, such as `plot`.  The convention is to import it as `plt`:

    import matplotlib.pyplot as plt
    
You can lauch IPython with `matplotlib` integration by typing:

    ipython --pylab
    
    ipython notebook --pylab=inline
    
These two methods are the best ways to use matplotlib interactively.  When you do this, you will have access to several `numpy` and `pyplot` in the same namespace.

###`pylab`

```python
scatter(arange(50), randn(50))
show()
```

```python
plot(randn(50), color='red')
show()
```

###`pyplot`

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.scatter(np.arange(50), np.random.randn(50))
plt.show()
```

## Figures and Subfigures

First, we create a blank figure.  Then we add two subpots, both on the same row.

```python
fig = plt.figure()

ax1 = fig.add_subplot(1,2,1)
ax1.plot(randn(50), color='red')

ax2 = fig.add_subplot(1,2,2)
ax2.scatter(np.arange(50), randn(50))
plt.show()
```

## The `plt.subplots()` command

```python
fig, ax = plt.subplots(2,3)

ax[0,1].plot(randn(50), color='green', linestyle='-')
ax[1,2].scatter(np.arange(50), randn(50), color='red')
ax[1,0].hist(randn(50))
plt.show()
```

## The `subplot2grid` command

```python
fig = plt.figure()
ax1 = plt.subplot2grid((3,3), (0,0), colspan=3)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2)
ax3 = plt.subplot2grid((3,3), (1,2), rowspan=2)
ax4 = plt.subplot2grid((3,3), (2,0))
ax5 = plt.subplot2grid((3,3), (2,1))
fig.tight_layout()
```

```python
subplots( 3, 1, sharex = True)
show()
```

```python
fig, axes = plt.subplots( 2, 2, sharex = True, sharey = True)
plt.subplots_adjust( wspace = 0.05, hspace = 0.1)
```

## Styles

```python
lines_types = ['-','-.', ':', 'steps']
lines_marker = ['+', 'o', '*', 's', ',', '.', '1']
```

```python
x = np.linspace(0,5,10)

# Figsize is in inches
subplots(1,1, figsize=(9,6))

for i, l in enumerate(lines_types):
    plot(x, x+i, ls=l, lw=2)

for j, l in enumerate(lines_marker):
    plot(x, x+i+j, ls='*', marker=l, markersize=6)  
```

##Basic plots

```python
import os
import pandas as pd
```

```python
cars = pd.read_csv(os.path.join('data','cars.csv'))
lenbefore = len(cars)
cars = cars.dropna()
lenafter = len(cars)
print '%s records reduced to %s' % (lenbefore, lenafter)
names = pd.read_csv(os.path.join('data','names.csv'))
iris = pd.read_csv(os.path.join('data','iris.csv'))
```

###Scatter

```python
plot(cars['wt'], cars['mpg'], 'o', alpha = .5, color = 'green') # the 'o' is for points
xlabel('weight')
ylabel('efficiency mpg')
title("Weight vs Efficiency for cars in the 70's and 80's")
show()
```

```python
c4 = cars[cars['cyl'] == 4]
c6 = cars[cars['cyl'] == 6]
c8 = cars[cars['cyl'] == 8]

plot(c4['wt'], c4['mpg'], 'o', label='4 cyl', alpha = .5, color = 'Green')
plot(c6['wt'], c6['mpg'], 'o', label='6 cyl', alpha = .5, color = 'Yellow')
plot(c8['wt'], c8['mpg'], 'o', label='8 cyl', alpha = .5, color = 'Red')
legend() #if you label
show()
```

###Line

```python
plot(names['year'], names['M'])
plot(names['year'][::4], names['F'][::4],'-o', alpha=0.5)
show()
```

```python
from scipy import stats

x = cars['wt'].values
y = cars['mpg'].values
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)

ye = slope*x + intercept
```

```python
plot(c4['wt'], c4['mpg'], 'o', label='4 cyl')
plot(c6['wt'], c6['mpg'], 'o', label='6 cyl')
plot(c8['wt'], c8['mpg'], 'o', label='8 cyl')
plot(x, ye, label='linear fit')
legend()
show()
```

### Compare distributions

```python
x1 = 100 + 15 * np.random.randn(100)
x2 = 50 + 10 * np.random.randn(100)

density1 = stats.kde.gaussian_kde(x1)
density2 = stats.kde.gaussian_kde(x2)

x = np.linspace(0, 150, 200)
```

```python
fill_between(x, 0, density1(x), alpha=0.5, color='#1B9E77')
fill_between(x, 0, density2(x), alpha=0.5, color='#D95F02')

plot([100,100],[0,density1(100)], color ='#1B9E77', linewidth=1, linestyle="--")
plot([50,50],[0,density2(50)], color ='#D95F02', linewidth=1, linestyle="--")
show()
```

###Stacked bar

Ref: [matplotlib examples](http://matplotlib.org/examples/pylab_examples/bar_stacked.html)

```python
N = 5
menMeans   = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)

ind = np.arange(N)+0.1    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

bar(ind, menMeans,   width, color='#1B9E77', label='men')
bar(ind, womenMeans, width, color='#D95F02', bottom=menMeans, label='women')

ylabel('Scores')
title('Scores by group and gender')
xticks(ind+width/2., ('G1', 'G2', 'G3', 'G4', 'G5') )
yticks(np.arange(0,81,10))
legend()
show()
```
