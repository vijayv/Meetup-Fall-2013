```python
import style
```

From http://messymind.net/2012/07/making-matplotlib-look-like-ggplot/

```python
c4 = cars[cars['cyl'] == 4]
c6 = cars[cars['cyl'] == 6]
c8 = cars[cars['cyl'] == 8]

fig, ax = plt.subplots()

ax.scatter(c4['wt'], c4['mpg'], s=50, c='#FF7F00', edgecolors='none', label='4 cyl', alpha = .75)
ax.scatter(c6['wt'], c6['mpg'], s=50, c='#4DAF4A', edgecolors='none', label='6 cyl', alpha = .75)
ax.scatter(c8['wt'], c8['mpg'], s=50, c='#377EB8', edgecolors='none', label='8 cyl', alpha = .75)

ax.set_xlabel('wt')
ax.set_ylabel('mpg')

ax.legend(loc='best', scatterpoints=1)

style.ggplot(ax)
show()
```

##Change the defaults in `matplotlibrc`

You can also specify most of this in the `matplotlibrc` file.  Here is an example from [huynf@github](https://gist.github.com/huyng/816622).

        axes.color_cycle    : 348ABD, 7A68A6, A60628, 467821, CF4457, 188487, E24A33
        axes.facecolor      : eeeeee   # axes background color
        axes.edgecolor      : bcbcbc   # axes edge color
        axes.linewidth      : 1        # edge linewidth
        axes.grid           : True     # display grid or not
        axes.titlesize      : x-large  # fontsize of the axes title
        axes.labelsize      : large    # fontsize of the x any y labels
        axes.labelcolor     : 555555
        axes.axisbelow      : True  
        patch.linewidth        : 0.5     # edge width in points
        patch.facecolor        : 348ABD # blue
        patch.edgecolor        : eeeeee
        patch.antialiased      : True    # render patches in antialised (no jaggies)
        xtick.major.size     : 0      # major tick size in points
        xtick.minor.size     : 0      # minor tick size in points
        xtick.major.pad      : 6      # distance to major tick label in points
        xtick.minor.pad      : 6      # distance to the minor tick label in points
        xtick.color          : 555555      # color of the tick labels
        xtick.direction      : in     # direction: in or out
        ytick.major.size     : 0      # major tick size in points
        ytick.minor.size     : 0      # minor tick size in points
        ytick.major.pad      : 6      # distance to major tick label in points
        ytick.minor.pad      : 6      # distance to the minor tick label in points
        ytick.color          : 555555      # color of the tick labels
        ytick.direction      : in     # direction: in or out
        figure.facecolor : 0.85    # figure facecolor; 0.75 is scalar gray
        figure.edgecolor : 0.50
        figure.figsize   : 11, 8 
        legend.fancybox  : True

## On the fly with `mpl`

```python
import matplotlib as mpl
mpl.rcParams['figure.edgecolor'] = 'grey'
mpl.rcParams['figure.facecolor'] = 'white'
mpl.rcParams['figure.figsize'] = '12, 6'
mpl.rcParams['axes.facecolor'] = '#eeeeee'
mpl.rcParams['axes.edgecolor'] = '#bcbcbc'
mpl.rcParams['axes.linewidth'] = '1' 
mpl.rcParams['axes.labelcolor'] = '#555555'
mpl.rcParams['axes.grid'] = 'True' 
mpl.rcParams['grid.color'] = '0.95' 
mpl.rcParams['grid.linestyle'] = '-'
mpl.rcParams['grid.linewidth'] = '2'
mpl.rcParams['ytick.color'] = '#555555' 
mpl.rcParams['xtick.color'] = '#555555' 
mpl.rcParams['axes.axisbelow'] = 'True'
mpl.rcParams['axes.color_cycle'] = '348ABD, 7A68A6, A60628, 467821, CF4457, 188487, E24A33'
mpl.rcParams['legend.fancybox'] = 'True'
```

```python
plot(names['year'], names['T'], linewidth=1.0, label='Total')
plot(names['year'], names['M'], linewidth=1.0, label='Male')
title('Births per year')
legend(loc='upper left')
show()
```
